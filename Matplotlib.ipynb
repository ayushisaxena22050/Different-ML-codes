```python
import numpy as np
t= np.arange(0.,5.,0.2)
plt.plot(t,t**2,'b--',label ='^2')
plt.plot(t,t**2.2,'rs',label='^2.2')
plt.plot(t,t**2.5,'g^',label='^2.5')
plt.grid()
plt.legend()
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1= np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)
plt.figure(1)
plt.subplot(211)
plt.grid()
plt.plot(t1,f(t1),'b-')
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2),'r--')
plt.show()
```

```python
url='https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
```

```python
data=pd.read_csv(url,sep=',',header=None)
```

```python
data.shape
```

```python
data.columns=['sepal_length','sepal_width','petal_length','petal_width','species']
```

```python
data['species'].value_counts()
```

## 2D Scatter plot

```python
data.plot(kind='scatter',x='sepal_length',y='sepal_width',grid=True)
plt.show() 
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(data,hue="species",size=4) \
.map(plt.scatter, "sepal_length","sepal_width") \
.add_legend()
```

using sepal_length and sepal width feature we can distinguish between setosa from other flowers.


seperating virginia from versicolor is much harder as they are linearly inseperable.

```python
plt.close()
sns.set_style('whitegrid')
sns.pairplot(data,hue="species",size = 3)
plt.show()
```
